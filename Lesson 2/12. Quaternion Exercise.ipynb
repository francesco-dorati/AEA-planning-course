```python
import numpy as np
```

```python
def quat_to_eul(quat):
  a = quat[0]
  b = quat[1]
  c = quat[2]
  d = quat[3]

  roll = np.arctan(2*(a*b+c*d)/(1-2*(b*b + c*c)))
  pitch = np.arcsin(2*(a*c-d*b))
  yaw = np.arctan(2*(a*d+c*b)/(1-2*(d*d + c*c)))
  return [roll, pitch, yaw]
```

```python
def eul_to_quat(eul):
  # angles in rad
  roll = eul[0]
  pitch = eul[1]
  yaw = eul[2]

  a = np.cos(roll/2)*np.cos(pitch/2)*np.cos(yaw/2) + np.sin(roll/2)*np.sin(pitch/2)*np.sin(yaw/2)
  b = np.sin(roll/2)*np.cos(pitch/2)*np.cos(yaw/2) - np.cos(roll/2)*np.sin(pitch/2)*np.sin(yaw/2)
  c = np.cos(roll/2)*np.sin(pitch/2)*np.cos(yaw/2) + np.sin(roll/2)*np.cos(pitch/2)*np.sin(yaw/2)
  d = np.cos(roll/2)*np.cos(pitch/2)*np.sin(yaw/2) - np.sin(roll/2)*np.sin(pitch/2)*np.cos(yaw/2)
  return [a, b, c, d]
```

```python

euler = np.array([np.deg2rad(90), np.deg2rad(30), np.deg2rad(0)])

q = eul_to_quat(euler) 
print(q)

# should be [ 1.570  0.523  0.]
e = quat_to_eul(q)
print(e)
```
